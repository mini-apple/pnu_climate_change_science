# 1. 세계 기상 관측소 plotting

```python
import pandas as pd
import numpy as np

data = pd.read_csv('C:/Users/UICHEOL/Desktop/기후변화과학/HW1/practice_data.csv')

data
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
from matplotlib import style

style.use("seaborn-talk")
m = Basemap(projection="mill")
m.drawcoastlines()
m.drawcountries()

for i in data.index :
    x, y = m(data['lon'][i], data['lat'][i])
    m.plot(x, y,'bo', ms=0.5)     

plt.show()
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style

style.use("seaborn-talk")
m = Basemap(projection="robin", resolution="i", area_thresh=1000.0, lat_0=0, lon_0=0)
m.drawcoastlines()
m.drawcountries()
#m.fillcontinents(color="gray")
m.drawmapboundary()
m.drawmeridians(np.arange(0,360,30))
m.drawparallels(np.arange(-90,90,30))

for i in data.index :
    x, y = m(data['lon'][i], data['lat'][i])
    m.plot(x, y,'bo', ms=0.5)  

plt.show()
```
